# Loading the location_list.txt file

#### There are mostly **grids** in the location_list.txt, so we get rid of them and only consider the named locations (cities)

```python
location_list={"names":[],"locations":[],"lat":[], "lon":[]}
c=0
with open("location_list.txt", "r") as file:
    for line in file.readlines():
        name, id, lat, lon = line.strip().split("\t")
        # Stop when we reach the grids
        if name.startswith("grid"):
            print(f"Number of named locations: {c}")
            print(f"First grid id: {id}")
            break
        location_list["names"].append(name)
        location_list["locations"].append(int(id))
        location_list["lat"].append(float(lat))
        location_list["lon"].append(float(lon))
        c += 1
        
# Converting the dictionary to a DataFrame
location_list_df = pd.DataFrame(location_list)
location_list_df
```

```python
location_list_df.to_csv('named_location_list.csv', index=False)
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(location_list_df['lon'], location_list_df['lat'], s=1, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Named Locations on World Map')
# plt.grid(True)
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.tight_layout()
plt.show()
```

# Considering a subset of cities which have 3D tile support on google maps

```python
import pandas as pd
data = pd.read_csv('sea_level_change_ssp585_medium_confidence_reduced_locations.csv')
filtered_city_names=pd.read_csv('filtered_100_cities.csv')
filtered_city_names.head()
```

```python
filtered_city_names.info()
```

```python
filtered_data =  pd.merge(
    filtered_city_names,
    data,
    on=["names"],   # shared columns
    how="inner"                       
)

filtered_data.head()
```

```python
filtered_data.info()
```

```python
# Row counts per city name (sorted by count)
name_counts = (
    filtered_data['names']
    .value_counts()
    .reset_index()
    .rename(columns={'index': 'names', 'names': 'row_count'})
)
name_counts.head(20)
```

### There are 3 instances of VENEZIA and 2 instances of NAPOLI, ALICANTE,SYDNEY,PALERMO we need to consider only one instance for each

```python
# Keep only first 1498 rows for VENEZIA and NAPOLI
mask = filtered_data.groupby('names').cumcount() < 1498
filtered_data = filtered_data[mask | ~filtered_data['names'].isin(['VENEZIA', 'NAPOLI','ALICANTE','SYDNEY','PALERMO'])]
filtered_data.reset_index(drop=True, inplace=True)
filtered_data.info()
```

```python
# Row counts per city name (sorted by count)
name_counts = (
    filtered_data['names']
    .value_counts()
    .reset_index()
    .rename(columns={'index': 'names', 'names': 'row_count'})
)
name_counts
```

```python
filtered_data.to_csv('final_filtered_data.csv', index=False)
```

# Considering only the median values

```python
filtered_data_median = filtered_data[filtered_data['quantiles'] == 0.5]
filtered_data_median.drop(columns=['quantiles'], inplace=True)
filtered_data_median.reset_index(drop=True, inplace=True)
filtered_data_median.info()
```
